```python
import string
from collections import Counter
import matplotlib.pyplot as plt
```

```python
import GetOldTweets3 as got
def tweets():
    criteria=got.manager.TweetCriteria().setQuerySearch('coronavirus')\
    .setSince('2020-01-22')\
    .setUntil('2020-02-22')\
    .setMaxTweets(200)
    tweets=got.manager.TweetManager.getTweets(criteria)
    text_tweets=[[tweet.text] for tweet in tweets]
    return text_tweets
    
```

```python
tweets()
```

```python
text=''
text_tweets=tweets()
leng=len(text_tweets)
for i in range(0,leng):
    text=text_tweets[i][0]+' '+text
print(text)
```

```python
lower_case = text.lower()
```

```python
from string import punctuation
```

```python
print(punctuation)
```

```python
cleaned_text=[word for word in text.split() if word not in punctuation]
```

```python
print(cleaned_text)
```

```python
from nltk.corpus import stopwords
stopwords=stopwords.words('english')
```

```python
stopwords
```

```python
final_text=[word for word in cleaned_text if word not in stopwords]
texts=''
for i in range(len(final_text)):
    texts=texts+' '+final_text[i]
print(texts)
```

```python
emotion_list = []
with open('emotions.txt', 'r') as file:
    for line in file:
        clear_line = line.replace('\n', '').replace(',', '').replace("'", '').strip()
        word, emotion = clear_line.split(':')
        if word in texts:
            emotion_list.append(emotion)

w = Counter(emotion_list)
print(w)

```

```python
fig,ax1=plt.subplots()
ax1.bar(w.keys(),w.values())
fig.autofmt_xdate()
plt.show()
```
